````python
# 3. Carregamento dos dados
try:
    # ATENÇÃO: Verifique se o nome do arquivo está correto no seu Google Colab
    df = pd.read_excel('Lab 05 - EvasaoClientesClassificados.xlsx')
    print("✅ Dados de treinamento 'Lab 05 - EvasaoClientesClassificados.xlsx' carregados!")

    # Preparando o contexto para a IA
    contexto_colunas = df.dtypes.to_string()
    amostra_dados = df.head(3).to_markdown()
except Exception as e:
    print(f"❌ Erro: Certifique-se de subir o arquivo Excel. Detalhe: {e}")
````

````python
# 4. FUNÇÃO DE AUTO-ML (O Agente que constrói a IA de Predição)
def executar_automl_nativo(pergunta_negocio="Crie um modelo preditivo"):
    """
    Esta função usa o Gemini para escrever e executar um pipeline completo
    de Machine Learning usando Scikit-Learn.
    """
    # Utilizando o modelo Gemini 3 Flash Preview para acesso Free Tier
    model = genai.GenerativeModel('gemini-3-flash-preview')

    prompt = f"""
    Você é um Cientista de Dados Sênior especializado em Scikit-Learn.
    O dataframe 'df' já está carregado e possui as colunas: {contexto_colunas}
    A coluna alvo (Target) para prever é 'Perdido?'.

    SUA TAREFA:
    Escreva um código Python completo que:
    1. Remova colunas irrelevantes (como ID Cliente).
    2. Trate variáveis categóricas (como Sexo, Região, Nivel Cobertura) usando LabelEncoder ou dummies.
    3. Divida os dados em treino e teste (80/20).
    4. Treine um modelo de Machine Learning (Random Forest por padrão, a menos que pedido diferente).
    5. Imprima a Acurácia e o Relatório de Classificação.
    6. Gere um gráfico de importância das variáveis (Feature Importance) usando matplotlib ou seaborn.
    7. RESPONDA À PERGUNTA: {pergunta_negocio}

    REGRAS:
    - Retorne APENAS o código puro dentro de blocos ```python.
    - Não explique nada fora do código.
    - Use 'plt.show()' para exibir o gráfico.
    """

    try:
        response = model.generate_content(prompt)
        # Extrai o código de dentro do bloco markdown
        codigo = response.text.split("```python")[1].split("```")[0].strip()

        print(f"🤖 Agente de IA iniciando pipeline de Machine Learning...\n")
        # Executa o código gerado no contexto global
        exec(codigo, globals())
    except Exception as e:
        print(f"⚠️ Erro no processo de AutoML: {e}")
````

````python
# 5. EXEMPLOS DE USO:
executar_automl_nativo("Identifique as 3 variáveis que mais influenciam a saída do cliente.")
````
